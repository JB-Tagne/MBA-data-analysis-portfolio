```python
# Étape 1: Demander à l'utilisateur le montant en euros

# Utilisation d'une boucle pour s'assurer que l'entrée est un nombre valide
while True:
    try:
        montant_euros_str = input("Veuillez entrer le montant en euros à convertir : ")
        # Remplacer la virgule par un point si l'utilisateur saisit avec une virgule
        montant_euros_str = montant_euros_str.replace(',', '.')
        montant_euros = float(montant_euros_str)
        if montant_euros < 0:
            print("Le montant ne peut pas être négatif. Veuillez réessayer.")
        else:
            break # Sortir de la boucle si l'entrée est valide
    except ValueError:
        print("Entrée invalide. Veuillez entrer un nombre (par exemple, 100 ou 50.75).")

# Étape 2: Obtenir des taux de change
# Pour cet exemple, nous allons utiliser un taux de change fixe.
# Au moment de l'écriture (vérifiez toujours les taux actuels pour des transactions réelles !)
# 1 EUR ≈ 1.08 USD (ceci est un exemple, le taux fluctue)
taux_eur_usd = 1.08

# Optionnel: Pour obtenir des taux en temps réel (plus complexe, nécessite une bibliothèque et une clé API)
# Vous pourriez utiliser une API comme celle de 'ExchangeRate-API' ou 'Open Exchange Rates'.
# Exemple conceptuel (ne fonctionnera pas sans la bibliothèque 'requests' et une clé API valide):
#
# import requests
# try:
#     # Remplacez YOUR_API_KEY par votre clé API réelle
#     # response = requests.get(f"https://v6.exchangerate-api.com/v6/YOUR_API_KEY/latest/EUR")
#     # data = response.json()
#     # if response.status_code == 200 and data.get("result") == "success":
#     #     taux_eur_usd_api = data['conversion_rates']['USD']
#     #     print(f"Taux de change actuel EUR/USD récupéré via API : {taux_eur_usd_api}")
#     #     taux_eur_usd = taux_eur_usd_api # Utiliser le taux de l'API
#     # else:
#     #     print("Impossible de récupérer le taux de change via l'API. Utilisation du taux fixe.")
#     #     # taux_eur_usd reste à la valeur fixe définie ci-dessus
# except Exception as e:
#     print(f"Erreur lors de la récupération du taux via API : {e}. Utilisation du taux fixe.")
#
# Note : L'utilisation d'API gratuites peut avoir des limitations (nombre de requêtes, fraîcheur des données).

print(f"\nUtilisation du taux de change : 1 EUR = {taux_eur_usd} USD")

# Étape 3: Faire le calcul
montant_dollars = montant_euros * taux_eur_usd

# Étape 4: Afficher le résultat
# Utilisation d'une f-string pour formater la sortie à deux décimales
print(f"\n--- Résultat de la Conversion ---")
print(f"{montant_euros:.2f} EUR équivalent à {montant_dollars:.2f} USD.")
print(f"-------------------------------")
```
